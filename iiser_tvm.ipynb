```python
import requests
from bs4 import BeautifulSoup

url= 'https://www.iisertvm.ac.in/#'


r = requests.get(url)

if r.status_code ==200:
    soup= BeautifulSoup(r.text, 'html.parser')

print(soup.prettify())
```

Accessing departments

```python
page_link = soup.find('ul', class_ = 'main_inner_sub_nav')

if page_link:
    links = page_link.find_all('a')
    for link in links:
        href = link.get('href')
        print(href)
```

Accessing individual faculty pages

```python
page_link = soup.find('ul', class_ = 'main_inner_sub_nav')

if page_link:
    links = page_link.find_all('a')
    for link in links:
        href = link.get('href')
    
        # Make a request to the individual page
        individual_page_response = requests.get(href)

        # Check if the request to the individual page was successful
        if individual_page_response.status_code == 200:
            # Parse the HTML content of the individual page
            individual_page_soup = BeautifulSoup(individual_page_response.text, 'html.parser')

            # Find <div> with class 'faculty name'
            faculty_name_divs = individual_page_soup.find_all('div', class_='faculty_name')

            # Iterate through each faculty_name_div
            for faculty_name_div in faculty_name_divs:
                # Extract and print the name
                faculty_name = faculty_name_div.get_text(strip=True)
                print(f"Faculty Name: {faculty_name}")

                # Find the link within the 'faculty name' div
                faculty_name_link = faculty_name_div.find('a')['href'] if faculty_name_div.find('a') else None
                print(f"Faculty Name Link: {faculty_name_link}")

                # Find <div> with class 'faculty designation' (assuming it's next to 'faculty name')
                faculty_designation_div = faculty_name_div.find_next('div', class_='faculty_designation')

                # Extract and print the designation
                if faculty_designation_div:
                    faculty_designation = faculty_designation_div.get_text(strip=True)
                    print(f"Faculty Designation: {faculty_designation}")

                print("\n")  # Add a separator between each faculty's information
```

Accessing personal information of faculty

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Replace 'url' with the URL of the page you're scraping
url = 'https://www.iisertvm.ac.in/#'

# Make a request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the <ul> element with class 'main_inner_sub_nav'
    ul_element = soup.find('ul', class_='main_inner_sub_nav')

    # Check if the <ul> element is found
    if ul_element:
        # Find all <a> (anchor) tags within the <ul> element
        links = ul_element.find_all('a')

        # Create a list to store the data
        data = []

        # Iterate through each faculty's information
        for link in links:
            href = link.get('href')
            individual_page_response = requests.get(href)

            if individual_page_response.status_code == 200:
                individual_page_soup = BeautifulSoup(individual_page_response.text, 'html.parser')

                faculty_name_divs = individual_page_soup.find_all('div', class_='faculty_name')

                for faculty_name_div in faculty_name_divs:
                    faculty_name = faculty_name_div.get_text(strip=True)
                    faculty_name_link = faculty_name_div.find('a')['href'] if faculty_name_div.find('a') else None

                    faculty_designation_div = faculty_name_div.find_next('div', class_='faculty_designation')
                    faculty_designation = faculty_designation_div.get_text(strip=True) if faculty_designation_div else None

                    # Get the text of the link from the 'href' attribute
                    link_text = link.get_text(strip=True)

                    # Check if faculty_name_link is a valid URL
                    if faculty_name_link and faculty_name_link.startswith('http'):
                        # Iterate over each faculty's personal page link
                        href_personal_page = faculty_name_link
                        personal_page_response = requests.get(href_personal_page)

                        if personal_page_response.status_code == 200:
                            personal_page_soup = BeautifulSoup(personal_page_response.text, 'html.parser')

                            # Extract additional information from the individual faculty page
                            faculty_page_content_div = personal_page_soup.find(class_='faculty_page_contact')

                            # Extract faculty personal page link
                            try:
                                faculty_personal_page_svg = faculty_page_content_div.find(class_='fa-globe')
                                faculty_personal_page_link = faculty_personal_page_svg.find_next('a')['href'] if faculty_personal_page_svg and faculty_personal_page_svg.find_next('a') else None
                            except AttributeError:
                                faculty_personal_page_link = None

                            # Append the data to the list
                            data.append({
                                'Faculty Name': faculty_name,
                                'Designation': faculty_designation,
                                'Department': link_text,  # Assuming link_text is the department
                                'Personal Page Link': faculty_personal_page_link
                            })

        # Create a DataFrame from the list
        df = pd.DataFrame(data)

        # Display the DataFrame
        print(df)


    else:
        print("No <ul> element with class 'main_inner_sub_nav' found.")
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")


# Export the DataFrame to a CSV file
csv_filename = 'IISER_TVM_I.csv'
df.to_csv(csv_filename, index=False)
print(f"Data has been exported to {csv_filename}")
```

Accessing additional information

```python
import csv
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Function to extract information from a faculty page
def extract_faculty_info(url):
    try:
        if url.startswith('javascript:'):
            print(f"Skipping unsupported URL: {url}")
            return None

        response = requests.get(url)
        response.raise_for_status()  # Raise an HTTPError for bad responses

        soup = BeautifulSoup(response.text, 'html.parser')

        # Extracting information
        name_element = soup.find('div', class_='faculty_page_name')
        designation_element = soup.find('div', class_='faculty_page_designation')
        contact_element = soup.find('div', class_='faculty_page_contact')

        # Initialize variables
        name = designation = contact = website = ""

        # Checking if all required elements are found
        if name_element:
            name = name_element.text.strip()

        if designation_element:
            designation = designation_element.text.strip()

        if contact_element:
            contact_div = contact_element.find('div')
            contact = contact_div.text.strip() if contact_div else ""

        website_a = contact_element.find_all('div')[2].find('a') if contact_element and len(contact_element.find_all('div')) > 2 else None
        website = website_a['href'].strip() if website_a and 'href' in website_a.attrs else ""

        return name, designation, contact, website

    except requests.exceptions.RequestException as e:
        print(f"Error processing {url}: {e}")
        return "", "", "", ""
    except Exception as e:
        print(f"Unexpected error processing {url}: {e}")
        return "", "", "", ""


# Function to process CSV file and create a DataFrame
def process_csv(csv_file):
    data = []
    with open(csv_file, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            faculty_link = row['Faculty Link']
            if faculty_link:
                faculty_info = extract_faculty_info(faculty_link)
                if faculty_info:
                    data.append(list(faculty_info))

    # Creating a DataFrame
    columns = ['Name', 'Designation', 'Contact', 'Website']
    df = pd.DataFrame(data, columns=columns)
    return df

# Replace 'your_csv_file.csv' with the actual name of your CSV file
faculty_dataframe = process_csv(r"C:\Users\Sirsha\Desktop\IDC\project\DATA\IISER_TVM_I.csv")

# Display the DataFrame
print(faculty_dataframe)

# Save the DataFrame to a new CSV file
faculty_dataframe.to_csv('IISER_TVM_extended.csv', index=False)
```

STORING IN SQLITE

```python
import pandas as pd
import sqlite3

# Replace 'your_file.csv' and 'your_database.db' with the actual paths/names
csv_file_path = r"C:\Users\Sirsha\Desktop\IDC\project\IISER_TVM_I.csv"
db_file_path = r"C:\Users\Sirsha\Desktop\IDC\project\IISER_TVM_I.db"

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# Create a SQLite connection and cursor
conn = sqlite3.connect(db_file_path)
cursor = conn.cursor()

# Write the DataFrame to a SQLite table
df.to_sql('faculty_data', conn, index=False, if_exists='replace')

# Commit changes and close the connection
conn.commit()
conn.close()

print(f"CSV data has been successfully written to {db_file_path} in the 'faculty_data' table.")
```
