```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace 'your_file.csv' with the actual path to your CSV file
csv_file_path = r"C:\Users\Sirsha\Desktop\IDC\project\IISER_MOHALI_I.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# Assuming the column names are 'Faculty Name' and 'Position'
# You may need to adjust these names based on your actual column names
name_col = 'Faculty Name'
desig_col = 'Position'

# Group the DataFrame by the 'Position' column and count the number of faculty in each Position
desig_counts = df.groupby(desig_col).size()

# Combine Positions with frequency less than 5 into a single category 'Others'
threshold = 5
desig_counts_combined = desig_counts.copy()
desig_counts_combined['Others'] = desig_counts_combined[desig_counts_combined < threshold].sum()
desig_counts_combined = desig_counts_combined[desig_counts_combined >= threshold]

# Plot a pie chart for Position counts
plt.pie(desig_counts_combined, labels=desig_counts_combined.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribution of Faculty Across Positions')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace 'your_file.csv' with the actual path to your CSV file
csv_file_path = r"C:\Users\Sirsha\Desktop\IDC\project\IISER_MOHALI_I.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# Assuming the column names are 'Faculty Name', 'Position', and 'Department'
# You may need to adjust these names based on your actual column names
name_col = 'Faculty Name'
desig_col = 'Position'
dept_col = 'Department'

# Group the DataFrame by both 'Position' and 'Department' columns
grouped_df = df.groupby([desig_col, dept_col]).size().unstack()

# Increase the size of the graph
fig, ax = plt.subplots(figsize=(10, 6))

# Plot a stacked bar graph for Position and department counts
ax = grouped_df.plot(kind='bar', stacked=True, colormap='viridis', ax=ax)
ax.set_xlabel('Position')
ax.set_ylabel('Number of Faculty')
ax.set_title('Number of Faculty in Each Position and Department')

# Create a separate legend outside the graph with smaller font
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
legend = ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), title='Legend', title_fontsize='12', prop={'size': 8})

plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels

# Save the legend as a separate image
legend_fig = legend.figure
legend_fig.savefig('legend.png', bbox_inches='tight')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace 'your_file.csv' with the actual path to your CSV file
csv_file_path = r"C:\Users\Sirsha\Desktop\IDC\project\IISER_MOHALI_I.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# Assuming the column names are 'Faculty Name', 'Position', 'Department'
# You may need to adjust these names based on your actual column names
name_col = 'Faculty Name'
desig_col = 'Position'
dept_col = 'Department'

# Group the DataFrame by both 'Position' and 'Department' columns and count faculty
grouped_df = df.groupby([desig_col, dept_col]).size().reset_index(name='Count')

# Increase the size of the graph
fig, ax = plt.subplots(figsize=(10, 6))

# Scatter plot with bubble size representing faculty count
bubble_plot = ax.scatter(
    grouped_df[desig_col],
    grouped_df[dept_col],
    s=grouped_df['Count'] * 20,  # Adjust the multiplier to control bubble size
    alpha=0.5,  # Set transparency for better visibility
    c=grouped_df['Count'],  # Color bubbles based on faculty count
    cmap='plasma',  # Choose a colormap
    edgecolors='black',  # Add black borders around bubbles for better contrast
)

# Set labels and title
ax.set_xlabel('Position')
ax.set_ylabel('Department')
ax.set_title('Bubble Chart - Faculty Distribution')

# Add a colorbar to indicate the count
cbar = plt.colorbar(bubble_plot, ax=ax)
cbar.set_label('Faculty Count')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Replace 'your_file.csv' with the actual path to your CSV file
csv_file_path = r"C:\Users\Sirsha\Desktop\IDC\project\IISER_MOHALI_I.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# Assuming the column names are 'Faculty Name', 'Position', 'Department'
# You may need to adjust these names based on your actual column names
name_col = 'Faculty Name'
desig_col = 'Position'
dept_col = 'Department'

# Group the DataFrame by both 'Position' and 'Department' columns and count faculty
grouped_df = df.groupby([desig_col, dept_col]).size().reset_index(name='Count')

# Create a treemap using plotly.express
fig = px.treemap(
    grouped_df,
    path=[desig_col, dept_col],
    values='Count',
    title='Treemap - Faculty Distribution',
    hover_data=['Count'],
    color='Count',
    color_continuous_scale='viridis',  # Choose a colormap
)

# Show the plot
fig.show()
```
