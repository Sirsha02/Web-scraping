```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd

# Replace 'url' with the URL of the page you're scraping
url = 'https://www.iiserkol.ac.in/web/en/people/faculty/faculty-all/#gsc.tab=0'

# Make a request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all <td> elements with the specified style
    faculty_entries = soup.find('div', class_="container content3-div cms-plugin cms-plugin-569")

    entries = faculty_entries.find_all('td')

    data_list = []

    for entry in entries:
        faculty_elements = entry.find('h4')

        link = faculty_elements.find('a')
        f_link = link['href']
        faculty_link = f_link

        # Find the <a> (anchor) tag within the <h4> tag
        name_a = faculty_elements.a
        # Extract the text content of the <a> tag
        faculty_name = name_a.get_text(strip=True)

        # Find the <br> tags within the <h4> tag
        dept_tags = faculty_elements.find_all('br')
        # Extract the text content following the first <br> tag
        if dept_tags:
            designation = dept_tags[0].next_sibling.strip()

        # Find the <b> tag with text "Department:"
        department_tag = faculty_elements.find('b', string='Department:')
        # Extract the text content following the <b> tag with text "Department:"
        if department_tag:
            department = department_tag.next_sibling.strip()

        # Find the <b> tag with text "Email:"
        email_tag = faculty_elements.find('b', string='Email:')
        # Extract the text content following the <b> tag with text "Email:"
        if email_tag:
            email = email_tag.next_sibling.strip()

        # Create a dictionary for each faculty member
        faculty_data = {
            'Faculty Name': faculty_name,
            'Faculty Link': faculty_link,
            'Designation': designation,
            'Department': department,
            'Email': email
        }

        data_list.append(faculty_data)

    # Convert the list of dictionaries to a DataFrame
    df = pd.DataFrame(data_list)

    # Export the DataFrame to a CSV file
    csv_filename = 'IISER_KOLKATA.csv'
    df.to_csv(csv_filename, index=False)

    print(f"Data has been exported to {csv_filename}")

```

ACCESSING FACULTY PAGES

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Load the CSV file
csv_path = r"C:\Users\Sirsha\Desktop\IDC\project\IISER_KOLKATA.csv"
df = pd.read_csv(csv_path)

# Define a function to scrape link from a given URL
def scrape_link(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            link_div = soup.find('div', class_="innerdiv col-md-6")
            link = link_div.find_next('a')['href'] if link_div.find_next('a') else None
            return link
    except Exception as e:
        print(f'error scraping link from {url}: {str(e)}')
    return None

# Define a function to scrape content from a given URL
def scrape_content(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            content = soup.find_all('div', class_='col-md-12 innerdiv')
            return ' '.join([c.get_text(separator=' ', strip=True) for c in content]) if content else None
    except Exception as e:
        print(f"Error scraping content from {url}: {str(e)}")
    return None

# Initialize a list to store data for the new DataFrame
data = []

# Iterate through each row in the CSV
for index, row in df.iterrows():
    faculty_name = row['Faculty Name']
    faculty_link = row['Faculty Link']
    link = 'https://www.iiserkol.ac.in/web/en/people/faculty' + faculty_link + '/#gsc.tab=0'

    # Skip rows with empty faculty links
    if pd.isnull(faculty_link):
        continue

    # Get content from the faculty link
    faculty_content = scrape_content(link)
    personal_page_link = scrape_link(link)

    # Add data to the list
    data.append([faculty_name, faculty_link, personal_page_link, faculty_content])

print(data)
# Create a new DataFrame
columns = ['Faculty Name', 'Faculty Link', 'Personal Page', 'Content']
result_df = pd.DataFrame(data, columns=columns)

# Save the result to a new CSV file
result_df.to_csv('IISER_Kol_extended.csv', index=False)
```
